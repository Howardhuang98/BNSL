# Structure learning for Asian network

```python
import networkx as nx
from dlbn import estimators
import pandas as pd
```

prepare data and initialize estimator.
the result are stored as "run_result.csv"

```python
data = pd.read_csv(r"../datasets/Asian.csv")
est = estimators.estimator(data,"Asian_result.csv")
```

est.run() begins learning process;
use nx.draw() to visualize result.

```python
result = est.run()
nx.draw(result,with_labels=True)
```
