# Structure learning for Asian network

```python
from dlbn import estimators
import pandas as pd
from dlbn.score import MDL_score
```

prepare data and initialize estimator.
the result are stored as "run_result.csv"

```python
data = pd.read_csv(r"../datasets/Asian.csv")
est = estimators.estimator(data,"Asian_result.csv")
```

est.run() begins learning process;
use est.show() to visualize result.

```python
est.run()
```

use est.show() to visualize result.

```python
est.show()
```

use heurisitc search to find optimal Bayesian network.

```python
est2 = estimators.HC_estimator(data,"Asian_result.csv")
est2.run(num_iteration=100,score_method=MDL_score,direction='down')
```

```python
est2.show()
```

Obviouly, dynamic program estimator has lower MDL score, better performance, however it uses more calculate time.
